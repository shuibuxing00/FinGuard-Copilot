# Generate FinGuard architecture diagram

This notebook creates `architecture.png` in the project root and verifies the output.

```python
import sys
from subprocess import run
from pathlib import Path

root = Path(r'd:/FinGuard Copilot')
output_path = root / 'architecture.png'

try:
    from PIL import Image, ImageDraw, ImageFont
except ImportError:
    run([sys.executable, '-m', 'pip', 'install', 'Pillow'], check=True)
    from PIL import Image, ImageDraw, ImageFont

img = Image.new('RGB', (1200, 860), (255, 255, 255))
draw = ImageDraw.Draw(img)

try:
    font = ImageFont.truetype('arial.ttf', 22)
    font_small = ImageFont.truetype('arial.ttf', 18)
except Exception:
    font = ImageFont.load_default()
    font_small = ImageFont.load_default()

boxes = [
    ((430, 40, 770, 120), '#1E88E5', 'Streamlit UI'),
    ((430, 170, 770, 250), '#43A047', 'LangChain Agent'),
    ((180, 340, 420, 420), '#F5F5F5', 'Tool: get_user_profile'),
    ((450, 340, 690, 420), '#F5F5F5', 'Tool: get_recent_transactions'),
    ((720, 340, 960, 420), '#F5F5F5', 'Tool: get_device_history'),
    ((450, 460, 690, 540), '#F5F5F5', 'Tool: search_compliance'),
    ((100, 620, 320, 700), '#E0E0E0', 'Splunk (mock data)'),
    ((440, 620, 660, 700), '#E0E0E0', 'OpenAI'),
    ((780, 620, 1000, 700), '#E0E0E0', 'Chroma'),
    ((260, 140, 940, 580), '#FDECEA', 'Security Layer:\nLLM Guard, RBAC, Audit Trail, Pseudonymization')
]

for bbox, fill, text in boxes:
    x1, y1, x2, y2 = bbox
    if text.startswith('Security Layer'):
        draw.rectangle(bbox, fill=fill, outline='#E53935', width=4)
        for i, line in enumerate(text.split('\n')):
            bbox_line = draw.textbbox((0, 0), line, font=font_small)
            w = bbox_line[2] - bbox_line[0]
            h = bbox_line[3] - bbox_line[1]
            draw.text((x1 + (x2 - x1 - w) / 2, y1 + 20 + i * 26), line, fill='#000000', font=font_small)
    else:
        draw.rectangle(bbox, fill=fill, outline='#000000', width=2)
        bbox_text = draw.textbbox((0, 0), text, font=font)
        w = bbox_text[2] - bbox_text[0]
        h = bbox_text[3] - bbox_text[1]
        draw.text((x1 + (x2 - x1 - w) / 2, y1 + (y2 - y1 - h) / 2), text, fill='#FFFFFF' if fill not in ['#F5F5F5', '#E0E0E0', '#FDECEA'] else '#000000', font=font)

arrow = lambda x1, y1, x2, y2: draw.line((x1, y1, x2, y2), fill='#000000', width=4)

arrow(600, 120, 600, 170)
arrow(600, 250, 600, 340)
arrow(300, 420, 300, 620)
arrow(570, 420, 570, 620)
arrow(840, 420, 840, 620)
arrow(570, 540, 570, 620)

for x, y in [(600, 170), (600, 340), (300, 620), (570, 620), (840, 620), (570, 620)]:
    draw.polygon([(x - 8, y - 12), (x + 8, y - 12), (x, y)], fill='#000000')

label = 'Security layer protects UI, Agent, Tools'
bbox_label = draw.textbbox((0, 0), label, font=font_small)
w = bbox_label[2] - bbox_label[0]
h = bbox_label[3] - bbox_label[1]
draw.text((420, 580), label, fill='#000000', font=font_small)

img.save(output_path)
print('Saved:', output_path)
print('Exists:', output_path.exists())
```
